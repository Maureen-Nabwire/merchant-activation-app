```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import (
    classification_report, confusion_matrix, accuracy_score,
    precision_score, recall_score, f1_score, roc_auc_score, roc_curve
)

from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import GradientBoostingClassifier
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

import joblib
```

```python
#Clean data
import pandas as pd
from scripts.cleaning import clean_accounts_df 

# Load raw dataset 
df= pd.read_csv('data/accounts Data.csv')

#load clean dataset
df_clean = clean_accounts_df(df)
print(f"Cleaned data: {df_clean.shape[0]} rows, {df_clean.shape[1]} columns")
```

```python
# Configure features & target
features = ['sector', 'referral_source', 'account_type', 'country']
target = 'conversion'  

X = df_clean[features]
y = df_clean[target]

# Define preprocessor: OneHotEncode categorical columns, pass others through
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), features)
    ],
    remainder='passthrough'  
)

#Display normalized class distribution for the training set.
#print("Train class distribution:\n", y_train.value_counts(normalize=True)) 
```

```python
#Train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42, stratify=y
)
#stratify=y maintains the same class distribution in both train and test sets.

#Output
print("Train shape:", X_train.shape, "Test shape:", X_test.shape)

#Display normalized class distribution for the training set.
print("Train class distribution:\n", y_train.value_counts(normalize=True)) 

norm_counts = y_train.value_counts(normalize=True)

#sns.countplot(x=norm_counts.index, y=norm_counts.values)
sns.barplot(x=norm_counts.index, y=norm_counts.values)
#ax[0].set_title("Conversion Rates")
#sns.barplot(x=norm_counts.index, y=norm_counts.values)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12,5))
norm_counts = y_train.value_counts(normalize=True)

sns.countplot(x=y_train, ax=ax[0])
ax[0].set_title("Train Class Counts")

sns.countplot(x=y_test, ax=ax[1])
ax[1].set_title("Test Class Counts")

plt.tight_layout()
plt.show()
```

### Class imbalance
There is severe class imbalance in the dataset
Algorithms may get biased towards the majority class and thus tend to predict output as the majority class.
Minority class observations look like noise to the model and are ignored by the model.
Imbalanced dataset gives misleading accuracy score.
* Conversion rates for 0(inactivated) and 1(activated) indicate class imbalance (majority class is 0).
* To combat and deal with these issues, the following will be applied.
* Models might predict 0 more often unless techniques like:
1. Class weighting
2. Oversampling (SMOTE)
4. Evaluation metrics like F1-score, ROC-AUC are applied.

```python
# Fit preprocessor on training set and transform both sets
X_train_trans = preprocessor.fit_transform(X_train)
X_test_trans = preprocessor.transform(X_test)

# Build feature names for interpretation 
try:
    # primary attempt: use OneHotEncoder's get_feature_names_out
    ohe = preprocessor.named_transformers_['cat']
    ohe_names = ohe.get_feature_names_out(features).tolist() #  generates names for encoded features.
    passthrough_cols = [c for c in X_train.columns if c not in features] #  keeps original columns that were not transformed.
    feature_names = ohe_names + passthrough_cols
except Exception:
    # fallback simple numeric index names
    feature_names = [f"f_{i}" for i in range(X_train_trans.shape[1])] # If the above fails, it assigns generic names like f_0, f_1, ....

print("Encoded train shape:", X_train_trans.shape)
```

* Train shape: (3409, 4 ): Original has 4 features
* Encoded train shape: (3409, 103): Encoded one has 103 features. This increase suggests:
* Your 4 categorical columns expanded into 103 binary features
â†’ This is normal because One-Hot Encoding creates multiple columns per category.
1. Multiple categorical variables were one-hot encoded.
2. Each categorical variable had many unique categories.
3. Consider dimensionality reduction.
        

### 1. Apply SMOTE on Training Dataset to habdle class imbalance

```python
# Apply SMOTE on training set 
# Synthetic Minority Oversampling Technique
sm = SMOTE(random_state=42)
X_resampled, y_resampled = sm.fit_resample(X_train_trans, y_train) #resample training data
# Works on the encoded training set (X_train_trans) and original labels (y_train).

print("After SMOTE:", X_resampled.shape)      
print("Resampled class distribution:\n", pd.Series(y_resampled).value_counts())
```

Fatures increased from 4 to 103.

```python
fig, ax = plt.subplots(1, 2, figsize=(12,5))

sns.countplot(x=y_resampled, ax=ax[0])
ax[0].set_title("Train Class Counts")

sns.countplot(x=y_test, ax=ax[1])
ax[1].set_title("Test Class Counts")

plt.tight_layout()
plt.show()
```

Before SMOTE: (3409, 103).
After SMOTE: (5580, 103) 5580 rows-> dataset size increased due to synthetic samples while features remained the same (103).

Class Distribution After SMOTE:
* conversion = 0: 2790 samples
* conversion = 1: 2790 samples
* Perfectly balanced classes (50/50). Improved model performance on minority class and consequently reduces risk of poor recall for minority class.

  For Test Set
You should not resample the test set
You should not apply SMOTE to the test set
You MUST evaluate on the original distribution
Otherwise your scores would be artificially inflated

### 2. Train the dataset

```python
#Train Decision Tree
dt_clf = DecisionTreeClassifier(
    max_depth=10,
    min_samples_split=5,
    min_samples_leaf=2,
    class_weight='balanced',
    random_state=42
)
# Limiting depth and minimum samples helps prevent overfitting, especially since SMOTE increased dataset size.
dt_clf.fit(X_resampled, y_resampled)
```

### 3. Evaluate performance using metrics like F1-score, ROC-AUC, and confusion matrix.

```python
#Evaluate Decision Tree (metrics + confusion matrix + ROC AUC)
# predictions of class labels and probabilities 
# Replace the following two lines with the correct variant above for your setup:
y_pred_dt = dt_clf.predict(X_test_trans)
y_prob_dt = dt_clf.predict_proba(X_test_trans)[:, 1] if hasattr(dt_clf, "predict_proba") else None

#classification report
print("Decision Tree â€” classification report")
print(classification_report(y_test, y_pred_dt))

#confusion matrix
cm = confusion_matrix(y_test, y_pred_dt)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Decision Tree â€” Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

#ROC-AUC
if y_prob_dt is not None:
    auc_dt = roc_auc_score(y_test, y_prob_dt)
    print(f"Decision Tree ROC AUC: {auc_dt:.4f}")
```

### Implications
* Model Performance
Accuracy is 72%, but this is misleading due to imbalance.
ROC-AUC of 0.6767 indicates moderate discrimination ability.
Class 1 (conversion) has poor precision and recall â†’ model struggles with minority class.

Why?
Even after SMOTE and class weights, Decision Tree may still overfit majority patterns.
Trees can be sensitive to noise and high dimensionality (103 features).
Use ensemble methods to improve model performance.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import plot_tree

rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    min_samples_leaf=1,
    min_samples_split=10,
    random_state=42
)
rf.fit(X_resampled, y_resampled)

# Extract one tree (e.g., the 5th)
estimator = rf.estimators_[5]

plt.figure(figsize=(24, 14))
plot_tree(
    estimator,
    #feature_names=X_train_trans,
    feature_names=feature_names,  # The list you built above
    class_names=['Not Activated', 'Activated'],
    filled=True,
    rounded=True,
    fontsize=10
)
plt.show()
```

```python
#Evaluate Random Forest (metrics + confusion matrix + ROC AUC)
# predictions of class labels and probabilities 
y_pred_rf = rf.predict(X_test_trans)
y_prob_rf = rf.predict_proba(X_test_trans)[:, 1] if hasattr(rf, "predict_proba") else None

#classification report
print("Random Forest â€” classification report")
print(classification_report(y_test, y_pred_rf))

#confusion matrix
cm = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Random Forest â€” Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

#ROC-AUC
if y_prob_dt is not None:
    auc_rf = roc_auc_score(y_test, y_prob_rf)
    print(f"Random Forest ROC AUC: {auc_rf:.4f}")
```

### 5. Train the ensemble method (Gradient Boosting) with the new parameters

```python

# Train Gradient Boosting (baseline)
gb_clf = GradientBoostingClassifier(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=5,
    random_state=42
)
gb_clf.fit(X_resampled, y_resampled)
```

### 6. Evaluate the gradient boosting method

```python

#Evaluate Gradient Boosting
y_pred_gb = gb_clf.predict(X_test_trans)
y_prob_gb = gb_clf.predict_proba(X_test_trans)[:, 1]

print("Gradient Boosting â€” classification report")
print(classification_report(y_test, y_pred_gb))

cm = confusion_matrix(y_test, y_pred_gb)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens')
plt.title("Gradient Boosting â€” Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

auc_gb = roc_auc_score(y_test, y_prob_gb)
print(f"Gradient Boosting ROC AUC: {auc_gb:.4f}")
```

### 7. Display ROC curves for both models

```python
# ROC curve for both models
plt.figure(figsize=(8,6))
if y_prob_dt is not None:
    fpr_dt, tpr_dt, _ = roc_curve(y_test, y_prob_dt)
    plt.plot(fpr_dt, tpr_dt, label=f'Decision Tree (AUC={roc_auc_score(y_test,y_prob_dt):.3f})')
fpr_gb, tpr_gb, _ = roc_curve(y_test, y_prob_gb)
plt.plot(fpr_gb, tpr_gb, label=f'Gradient Boosting (AUC={roc_auc_score(y_test,y_prob_gb):.3f})')

fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)
plt.plot(fpr_rf, tpr_rf, label=f'Random Forest (AUC={roc_auc_score(y_test,y_prob_rf):.3f})')
plt.plot([0,1], [0,1], 'k--', label='Chance')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves")
plt.legend()
plt.grid(True)
plt.show()
```

### 8. Display feaure importance for both models

```python
#Feature importances (top 30) for both models
def plot_top_feature_importances(model, feature_names, top_n=30, title=None):
    fi = pd.Series(model.feature_importances_, index=feature_names).sort_values(ascending=False).iloc[:top_n]
    plt.figure(figsize=(10, max(4, 0.25*len(fi))))
    sns.barplot(x=fi.values, y=fi.index)
    plt.title(title or "Feature importances")
    plt.xlabel("Importance")
    plt.tight_layout()
    plt.show()

# Decision Tree importances
plot_top_feature_importances(dt_clf, feature_names, top_n=30, title="Decision Tree â€” Top Feature Importances")

# Gradient Boosting importances
plot_top_feature_importances(gb_clf, feature_names, top_n=30, title="Gradient Boosting â€” Top Feature Importances")

# Random Forest importance
plot_top_feature_importances(rf, feature_names, top_n=30, title="Random Forest â€” Top Feature Importances")
```

### MODEL OPTIMIZATION
1. Hyperparameter tuning using cross validation

### Decision Tree

```python
# ----- 1. Decision Tree -----
dt_pipe = ImbPipeline(steps=[
    ('preprocess', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('clf', DecisionTreeClassifier(random_state=42))
])

dt_params = {
    'clf__max_depth': [5, 10, 15, None],
    'clf__min_samples_split': [2, 5, 10],
    'clf__min_samples_leaf': [1, 2, 5],
    'clf__criterion': ['gini', 'entropy'],
    'clf__class_weight': ['balanced']
}

# select the best set of hyperparameters based on your scoring metric (roc_auc).
dt_grid = GridSearchCV(dt_pipe, dt_params, cv=3, scoring='roc_auc', n_jobs=-1)
# Retrainn the model with the new hyperparameters
dt_grid.fit(X_train, y_train)
```

```python
#Evaluation of the new optimized model
y_pred = dt_grid.predict(X_test)
y_proba = dt_grid.predict_proba(X_test)[:, 1]#
#Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred))
#ROC-AUC score
roc_auc = roc_auc_score(y_test, y_proba)
print(f"ROC-AUC Score: {roc_auc:.4f}")
```

```python
# ----- 2. Random Forest -----
rf_pipe = ImbPipeline(steps=[
    ('preprocess', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('clf', RandomForestClassifier(random_state=42))
])

rf_params = {
    'clf__n_estimators': [200, 400],
    'clf__max_depth': [10, 20, None],
    'clf__min_samples_split': [2, 5, 10],
    'clf__min_samples_leaf': [1, 2, 5],
    'clf__class_weight': ['balanced']
}

rf_grid = GridSearchCV(rf_pipe, rf_params, cv=3, scoring='roc_auc', n_jobs=-1)
rf_grid.fit(X_train, y_train)
```

```python
#Evaluation of the new optimized model
y_pred = rf_grid.predict(X_test)
y_proba = rf_grid.predict_proba(X_test)[:, 1]#
#Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred))
#ROC-AUC score
roc_auc = roc_auc_score(y_test, y_proba)
print(f"ROC-AUC Score: {roc_auc:.4f}")
```

```python
# ----- 3. Gradient Boosting -----
gb_pipe = ImbPipeline(steps=[
    ('preprocess', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('clf', GradientBoostingClassifier(random_state=42))
])

gb_params = {
    'clf__n_estimators': [100, 200, 300],
    'clf__learning_rate': [0.05, 0.1],
    'clf__max_depth': [3, 5],
    'clf__subsample': [0.7, 1.0]
}

gb_grid = GridSearchCV(gb_pipe, gb_params, cv=3, scoring='roc_auc', n_jobs=-1)
gb_grid.fit(X_train, y_train)
```

```python
#Evaluation of the new optimized model
y_pred = gb_grid.predict(X_test)
y_proba = gb_grid.predict_proba(X_test)[:, 1]#
#Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred))
#ROC-AUC score
roc_auc = roc_auc_score(y_test, y_proba)
print(f"ROC-AUC Score: {roc_auc:.4f}")
```
